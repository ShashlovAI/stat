```python
import numpy as np
from scipy.stats import lognorm, norm, shapiro, kstest
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
np.random.seed (1)
lognorm_dataset = lognorm.rvs(s=0.5, scale=1, size=1000)
norm_dataset = r = norm.rvs(loc=0, scale=1, size=1000)
```

# Способ 1: гистограмма

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
ax[0].hist(lognorm_dataset, edgecolor='black', bins=20);
ax[1].hist(norm_dataset, edgecolor='black', bins=20);
```

# Способ 2: QQplot

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
sm.qqplot(lognorm_dataset, ax = ax[0], line='45')
sm.qqplot(norm_dataset, ax = ax[1], line='45')
plt.show()
```

# Способ 3: тест Шапиро-Уилка

```python
print(shapiro(lognorm_dataset))
print(shapiro(norm_dataset))
```

# Способ 4: тест Колмогорова-Смирнова

```python
print(kstest(lognorm_dataset, 'norm'))
print(kstest(norm_dataset, 'norm'))
```
